```python
class Model(nn.Module): #inherit nn.module class , base class
    def __init__(self,num_features):
      super().__init__() #inherit from parent class super is used
      self.linear = nn.Linear(num_features,1)
      self.sigmoid = nn.Sigmoid()

    def forward(self,features):
      out = self.linear(features) #wx+b
      out = self.sigmoid(out) #sig(z)
      return out
```

```python
#create a DS
features = torch.rand(10,5) #5 features = col
Model = Model(features.shape[1])

Model(features) #recommended syntax
#model.forward(features) can work but not recommended


#we get forward pass of all 10
```

<h1> HL NN 5 Input, 3 Neuron in HL, 1 output Neuron </h1>

```python
class Model(nn.Module): #inherit nn.module class , base class
    def __init__(self,num_features):
      super().__init__() #inherit from parent class super is used
      self.linear1 = nn.Linear(num_features,3)
      self.relu = nn.ReLU()
      self.linear2 = nn.Linear(3,1)
      self.sigmoid = nn.Sigmoid()

    def forward(self,features):
      out = self.linear1(features) #wx+b
      out = self.relu(out) #sig(z)
      out = self.linear2(out)
      out = self.sigmoid(out)
      return out
```

```python
features = torch.rand(10,5) #5 features = col
Model = Model(features.shape[1])

Model(features) #recommended syntax
```

<h2> Sequential Container </h2>

```python
class Model(nn.Module): #inherit nn.module class , base class
    def __init__(self,num_features):
      super().__init__() #inherit from parent class super is used
      self.network = nn.Sequential(
        nn.Linear(num_features,3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid()
      )

    def forward(self,features):
      out = self.network(features)
      return out
```

```python
features = torch.rand(10,5) #5 features = col
Model = Model(features.shape[1])

Model(features)
```

<h1> Change in pipeline code </h1>

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['Unnamed: 32','id'],inplace=True)
df.head()
```

```python
#test train split
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns=['diagnosis']),df['diagnosis'],test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
#
```

```python
#label encoding
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
x_train_tensor = torch.from_numpy(x_train).float()
x_test_tensor = torch.from_numpy(x_test).float()
y_train_tensor = torch.from_numpy(y_train).float()
y_test_tensor = torch.from_numpy(y_test).float()
```

```python
class MySimpleNN(nn.Module):

  def __init__(self,num_features):
    super().__init__()
    self.linear = nn.Linear(num_features,1)
    self.sigmoid = nn.Sigmoid()

  def forward(self,features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out

```

```python
learning_rate = 0.1
epochs = 50
```

```python
loss_function = nn.BCELoss()
```

```python
type(loss_function)
```

```python
model = MySimpleNN(x_train_tensor.shape[1])
```

```python
optimizer= torch.optim.SGD(model.parameters(),lr=learning_rate)
#model.parameters
```

```python
#create Model

#define loop
for epoch in range(epochs):
  #forward Pass
  y_pred = model(x_train_tensor)
  #print(y_pred)

  #loss calculate
  loss = loss_function(y_pred,y_train_tensor.view(-1,1)) #reshape data
  #print(loss)

  #prevent accumulation of gradient recommended to be before backward
  optimizer.zero_grad()

  #backward Pass

  loss.backward()
  #weight parameter update
  optimizer.step()


  #priint losss in each epoch
  print(f"Epoch: {epoch+1}, Loss: {loss.item()}")
```

```python
model.linear.weight
```

```python
 #model evaluation
with torch.no_grad():
  y_pred = model.forward(x_test_tensor)
  y_pred = (y_pred > 0.5).float()# prob to label
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f"Accuracy: {accuracy.item()}")

print(y_pred)
```
